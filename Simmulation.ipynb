## 1. Question

¿What are the negative effects on welfare when people make their consumption decisions without full knowledge of their preferences?


## 2. Motivation

Pension funds should be governed properly to ensure benefits from decision making with respect to investment and consumption decision at an older age. This is vulnerable when people base their decisions on wrong parameters. In that sense, we evaluate the welfare costs of suboptimal financial planning over the lifecycle, using the benchmark model of Bovenberg, Koijen, Nijman and Teulings (2007) [2](http://download.springer.com/static/pdf/602/art%253A10.1007%252Fs10645-007-9070-1.pdf?originUrl=http%3A%2F%2Flink.springer.com%2Farticle%2F10.1007%2Fs10645-007-9070-1&token2=exp=1485718515~acl=%2Fstatic%2Fpdf%2F602%2Fart%25253A10.1007%25252Fs10645-007-9070-1.pdf%3ForiginUrl%3Dhttp%253A%252F%252Flink.springer.com%252Farticle%252F10.1007%252Fs10645-007-9070-1*~hmac=1633fffab49b62656236f0586ae1372bc899de8f8a25e84f8ff052dea3280eef).  

In the Netherlands the most common form of pension system is the collective one. But currently few individual elements of choice have been introduced. It has been demonstrated that individual choice may be beneficial for people, if individual preferences are known. However, individual choice can make people worse off if people lacks of knowledge. We show how much the welfare loss is when individual preferences are not known.  

This study consists in two cases. The first case investigates suboptimal investments decisions as consequence of decision-making based on the wrong parameter of risk aversion ($\theta$). We check how overestimation or underestimation of risk aversion affects the welfare. The second case examines suboptimal investment decisions that result of decision making based on the wrong parameter of time preference ($\rho$). When time preference is higher than the optimal level, we can observe a consumption preference for the near future as opposed to the distant future and vice versa.  

## 3. Main asumptions
This section lays out our benchmark framework for exploring optimal life- time saving and investment (see also Merton (1971) [3](http://www.people.hbs.edu/rmerton/Optimum%20Consumption%20and%20Portfolio%20Rules.pdf), Merton and Samuelson (1974) [4](http://www.people.hbs.edu/rmerton/optionpricingwhenunderlingstock.pdf), Teulings and de Vries (2006) [5](http://arno.uvt.nl/show.cgi?fid=81734)). It describes the benchmark assumptions on financial markets, labor markets and preferences and discusses the parameter values employed in the numerical simulations.

### 3.1. Financial Markets
- A risk-free asset (a bond) is available.
- Equity-market risk is the only aggregate risk factor, which is traded through equity. Housing is abstracted from.
- The interest rate, ináation, the volatility of equity, and the equity risk premium are constant over time so that mean reversion and stochastic volatility are absent. These prices are not affected by the decisions of individuals or financial intermediaries. We thus take a partial-equilibrium perspective of a small open economy, which takes prices as given on the world market.
- Log stock returns are identically and independently distributed according to a normal distribution.
- Financial markets are dynamically complete if households and pension funds can continuously trade stocks and bonds without constraints. In some cases, households and pension funds face constraints.
- Death is predictable or perfect insurance of individual longevity risk is available. Aggregate longevity risk is thus absent.
- Whenever we consider the presence of collective pension schemes we assume that participants have access to the capital markets only through their pension scheme. From the householdís perspective markets are thus incomplete. Hence, individuals cannot offset the policies of their pension scheme by engaging in offsetting capital-market transactions. Consumption during the active period (i.e. when working) equals wage income minus the pension premium. Consumption in retirement is given by the pension benefit.

### 3.2. Labor markets
- The after-tax wage during the working career is constant and riskless with a fixed, exogenous retirement age. Human capital is thus paid out in the form of constant wage income until it is fully depleted at the age of retirement. Moreover, labor-market risks are absent.
- Labor supply is fixed.
- Wages are exogenous: pension premia thus reduce disposable incomes one for one.

### 3.3. Preferences
- Individuals aim to maximize lifetime utility, which is the weighted sum over time of expected utility at each point in time. Utility at a point in time depends only on consumption at that time. The weights of future expected utilities decline exponentially at the so called rate of time preference. Hence, people are impatient: at equal levels of consumption a marginal unit of future consumption adds less to utility than current consumption.  
- Preferences feature positive and constant relative risk aversion. Consumption in each contingency adds to utility. At larger levels of consumption, however, additional consumption adds less to utility. The negative elasticity of marginal utility with respect to the level of consumption is known as the coeficient of relative risk aversion. This measure for the sensitivity of marginal utility with respect to the level of consumption is constant and thus does not depend on the consumption level.  
Positive risk aversion implies that individuals have a taste for moderation across time and across contingencies. They prefer a smooth consumption level (with a low variance) rather than a highly volatile consumption stream (with a high variance) over time or across contingencies. More general models (than expected utility) distinguish the taste for moderation across contingencies from the taste for moderation across time. Risk aversion measures the taste for moderation across random outcomes. Highly risk-averse individuals feature a strong preference for stable consumption levels across various contingencies. They want their consumption level to be certain ("stable") irrespective of what happens (which is reflected in the costs of maintaining the consumption level in a particular contingency). The taste for moderation across time is related to the intertemporal elasticity of substitution. An individual exhibiting a low intertemporal elasticity of substitution prefers a stable level of consumption over time.
- Bequest motives are absent.
- Individuals start consuming when they enter the labor market.

In this section we are going to examine the effects of a misspecification of the optimal risk aversion parameter, $\theta$, and the optimal time preference parameter, $\rho$. The benchmark model of L.Bovenberg, R.Koijen, T.Nijman and C.Teulings [2](http://download.springer.com/static/pdf/602/art%253A10.1007%252Fs10645-007-9070-1.pdf?originUrl=http%3A%2F%2Flink.springer.com%2Farticle%2F10.1007%2Fs10645-007-9070-1&token2=exp=1485718515~acl=%2Fstatic%2Fpdf%2F602%2Fart%25253A10.1007%25252Fs10645-007-9070-1.pdf%3ForiginUrl%3Dhttp%253A%252F%252Flink.springer.com%252Farticle%252F10.1007%252Fs10645-007-9070-1*~hmac=1633fffab49b62656236f0586ae1372bc899de8f8a25e84f8ff052dea3280eef) is used and the true values for $\theta$ the risk aversion parameter, and $\rho$ the time preference parameter are equal to $\theta=5$ and $\rho=0.02$. The effect on consumption and financial wealth are investigated and compared to the true values.

### 4.2. Specification of model parameters

#### 4.2.1. Life cycle parameters

```python
D=60 #total time in pension fund
T=45 
deltat=1
start_age=20
```

#### 4.2.2. Economic parameters

```python
r=0.02 #interest rate
sigma=0.2 #volatility of risky assets
lambda_v= 0.04/sigma
```

#### 4.2.3. Specification of Age, Labor Earnings and Human Capital

```python
age=[]
for i in range(1,60,1):
    e=start_age+(i*deltat)
    age.append(e)

def Funcion_Y(i): #Labour earnings
    if age[i-1]<start_age+T:
        return 1
    else:
        return 0
    
def Funcion_H(i): #Human Capital
        e=(1/r)*(1-math.exp(-r*max(0,(start_age+T-age[i-1]))))
        return e

def Funcion_A(theta, rho):
    w=((1-theta)*r-rho)/theta+0.5*((1-theta)/theta**2)*lambda_v**2
    return w

def Funcion_g(i, theta, rho):
    e=(1/Funcion_A(theta, rho))*(math.exp(Funcion_A(theta, rho)*(D-(age[i-1]-start_age)))-1)
    return e

#Adding extra g, needed for certainty equivalence
def Funcion_g_ce(theta, rho):
    e=(1/Funcion_A(theta, rho))*(math.exp(Funcion_A(theta, rho)*D)-1)
    return e
```

### 4.3. Defining the paths of consumption, wealth and portfolio choices

The following functions define the dynamics of consumption, financial wealth and portfolio choices. These paths depend on the previously defined parameters and are found using recursive function calls, which are defined according to the appendix of L.Bovenberg, R.Koijen, T.Nijman and C.Teulings [2](http://download.springer.com/static/pdf/602/art%253A10.1007%252Fs10645-007-9070-1.pdf?originUrl=http%3A%2F%2Flink.springer.com%2Farticle%2F10.1007%2Fs10645-007-9070-1&token2=exp=1485718515~acl=%2Fstatic%2Fpdf%2F602%2Fart%25253A10.1007%25252Fs10645-007-9070-1.pdf%3ForiginUrl%3Dhttp%253A%252F%252Flink.springer.com%252Farticle%252F10.1007%252Fs10645-007-9070-1*~hmac=1633fffab49b62656236f0586ae1372bc899de8f8a25e84f8ff052dea3280eef).  

```python
def Funcion_F(n,theta,rho): #Financial Wealth
    if n==1:
        return 0
    else:
        for n in range(2,60,1):
            e=Funcion_F(n-1,theta,rho)+Funcion_F(n-1,theta,rho)*r*deltat+Funcion_fstar_times_F(n-1,theta,rho)*sigma*lambda_v*deltat+Funcion_fstar_times_F(n-1,theta,rho)*math.sqrt(deltat)*sigma*np.random.randn()+(Funcion_Y(n-1)-Funcion_C(n-1,theta,rho))*deltat
            Matrix=np.ones(50000)
            MatrixF=Matrix*e
            w=sum(MatrixF)/len(MatrixF)
            return w

def Funcion_C(n,theta,rho): #Consumption
    for n in range(1,60,1):
        e=(Funcion_F(n,theta,rho)+Funcion_H(n))*Funcion_g(n,theta,rho)**(-1)
        return e 
    
def Funcion_fstar_times_F(n,theta,rho):
    for n in range(1,60,1):
        e=(1/theta)*(lambda_v/sigma)*(Funcion_F(n,theta,rho)+Funcion_H(n))
        return e
    
def Funcion_fstar(n,theta,rho): #Portfolio choice (percentage of wealth invested in risky assets)
    if n==1:
        return float('nan')
    else:
        for n in range(2,60,1):
            e=Funcion_fstar_times_F(n,theta,rho)/Funcion_F(n,theta,rho)
            return e
```

### 5.1. Suboptimal Investment Decision that Result if Decision Making is based upon the Wrong Parameter of Relative Risk Aversion

#### 5.1.1. Effect on consumption

If a decision is made based on another value of the risk aversion we get a sub-optimal outcome. The effect on median consumption can be found in **Figure 1**.  

The higher the value of $\theta$, the more risk averse the individual. When analyzing the effect of a change in the risk aversion parameter in consumption, we ought to remember that the less risk averse the person is, the more risky choices of investment she will make, which will lead to higher profits on average. This result can be seen in **Figure 2**, the lower the risk aversion the higher the part in risky assets. This is why we can see that as we consider lower values for $\theta$, the median value of consumption rises. Also, because less averse people will make riskier and more profitable choices through time, the expected accumulated return will be higher the longer time we consider. So, for every value of $\theta$, the trend is growing upwards, but for low values of $\theta$ the growth rate of this trend is higher. It is also important to note that the variation of outcomes do increase as the risk aversion parameter decreases.  

If the average person in the pension fund has a lower risk aversion then the individual has (i.e. $\theta < 5$), the fund will invest more risky than the individual with $\theta = 5$ expects. Then we expect that the consumption will be lower than the average retiree what causes a welfare loss for the individual. The size of the welfare loss will be discussed in **section 6**.

  ### 6. Welfare Loss due to Wrong Parameter Values
In the current political debate about pensions in the Netherlands choice is an important topic. If choice is offered, participants can gain if their optimal preference parameters are different from the average participant. However the literature shows that participants don’t know their own preferences and that the architectural design of choice can influence the participants. Benartzi and Thaler [1](http://www.anderson.ucla.edu/faculty/shlomo.benartzi/autonomy.pdf) have studied the behavior regarding retirement investments of participants in defined contribution pension schemes. They have found that participants allocate their retirement savings dependent on the portfolio choices offered to them. If this results in a sub-optimal outcome, it may indicate welfare loss on individual level. In order to study the effect of wrong parameter specification, certainty equivalents are computed and compared to the optimal results of the benchmark model by Bovenberg et al. [2](http://download.springer.com/static/pdf/602/art%253A10.1007%252Fs10645-007-9070-1.pdf?originUrl=http%3A%2F%2Flink.springer.com%2Farticle%2F10.1007%2Fs10645-007-9070-1&token2=exp=1485718515~acl=%2Fstatic%2Fpdf%2F602%2Fart%25253A10.1007%25252Fs10645-007-9070-1.pdf%3ForiginUrl%3Dhttp%253A%252F%252Flink.springer.com%252Farticle%252F10.1007%252Fs10645-007-9070-1*~hmac=1633fffab49b62656236f0586ae1372bc899de8f8a25e84f8ff052dea3280eef). The results are shown in terms of welfare loss.  


$$E_t \left( \int_{t}^{D} \frac{e^{-\rho \left( s-t \right)}}{1-\theta} C_s^{1-\theta*}ds \right)=\int_{t}^{D} \frac{e^{-\rho \left( s-t \right)}}{1-\theta} ce_s^{1-\theta}ds$$

The certainty equivalent is computed using the following equation contained in the appendix (A5) of Bovenberg et al. [2](http://download.springer.com/static/pdf/602/art%253A10.1007%252Fs10645-007-9070-1.pdf?originUrl=http%3A%2F%2Flink.springer.com%2Farticle%2F10.1007%2Fs10645-007-9070-1&token2=exp=1485718515~acl=%2Fstatic%2Fpdf%2F602%2Fart%25253A10.1007%25252Fs10645-007-9070-1.pdf%3ForiginUrl%3Dhttp%253A%252F%252Flink.springer.com%252Farticle%252F10.1007%252Fs10645-007-9070-1*~hmac=1633fffab49b62656236f0586ae1372bc899de8f8a25e84f8ff052dea3280eef).

We computed the certainty equivalents for a consumption-investment policy. The certainty equivalent consumption level $ce$ follows from equation (A27) contained in the appendix of Bovenberg et al. Here $W_0$ is the wealth of the investor at period $0$. For further explanation on parameter $g$ we refer to the appendix of the paper of Bovenberg et al [2](http://download.springer.com/static/pdf/602/art%253A10.1007%252Fs10645-007-9070-1.pdf?originUrl=http%3A%2F%2Flink.springer.com%2Farticle%2F10.1007%2Fs10645-007-9070-1&token2=exp=1485718515~acl=%2Fstatic%2Fpdf%2F602%2Fart%25253A10.1007%25252Fs10645-007-9070-1.pdf%3ForiginUrl%3Dhttp%253A%252F%252Flink.springer.com%252Farticle%252F10.1007%252Fs10645-007-9070-1*~hmac=1633fffab49b62656236f0586ae1372bc899de8f8a25e84f8ff052dea3280eef).  

$$ce=\left(W_0^{1-\theta} g\left(D\right)^\theta \frac{\rho}{1-e^{\rho D}}\right)^{\frac{1}{1-\theta}}$$

We have computed the certainty equivalent for several parameters including the true ones, we can calculate the welfare loss. The welfare loss is equal to:  

$$\frac{ce^{true}-ce}{ce}$$

In the last equation, $ce^{true}$ is the certainty equivalent based on the true values of the parameters which are: $\rho=0.02$ and $\theta=5$. 

```python
#Creating equation A5 (Expected Utility during 20-80 years)

#as we esported vectors from an excel sheet, we turn them into arrays.
ct5r1=np.array(Ct5r1)
ct5r2=np.array(Ct5r2)
ct5r3=np.array(Ct5r3)
ct5r4=np.array(Ct5r4)
ct5r5=np.array(Ct5r5)

ct2r2=np.array(Ct2r2)
ct2r3=np.array(Ct3r2)
ct2r4=np.array(Ct4r2)
ct2r6=np.array(Ct6r2)
ct2r7=np.array(Ct7r2)
ct2r8=np.array(Ct8r2)
ct2r9=np.array(Ct9r2)
ct2r10=np.array(Ct10r2)

#creating equation A5 for different values of theta and rho: First we create the utility in each period
a5_t5r1=[]
a5_t5r2=[]
a5_t5r3=[]
a5_t5r4=[]
a5_t5r5=[]

a5_t2r2=[]
a5_t3r2=[]
a5_t4r2=[]
a5_t6r2=[]
a5_t7r2=[]
a5_t8r2=[]
a5_t9r2=[]
a5_t10r2=[]

for i in range(59):
    e1=(math.exp(-0.01*(age[i]-start_age))*(ct5r1[i])**(1-5))/(1-5)
    e2=(math.exp(-0.02*(age[i]-start_age))*(ct5r1[i])**(1-5))/(1-5)
    e3=(math.exp(-0.03*(age[i]-start_age))*(ct5r1[i])**(1-5))/(1-5)
    e4=(math.exp(-0.04*(age[i]-start_age))*(ct5r1[i])**(1-5))/(1-5)
    e5=(math.exp(-0.05*(age[i]-start_age))*(ct5r1[i])**(1-5))/(1-5)
    
    e6=(math.exp(-0.02*(age[i]-start_age))*(ct5r1[i])**(1-2))/(1-2)
    e7=(math.exp(-0.02*(age[i]-start_age))*(ct5r1[i])**(1-3))/(1-3)
    e8=(math.exp(-0.02*(age[i]-start_age))*(ct5r1[i])**(1-4))/(1-4)
    e9=(math.exp(-0.02*(age[i]-start_age))*(ct5r1[i])**(1-6))/(1-6)
    e10=(math.exp(-0.02*(age[i]-start_age))*(ct5r1[i])**(1-7))/(1-7)
    e11=(math.exp(-0.02*(age[i]-start_age))*(ct5r1[i])**(1-8))/(1-8)
    e12=(math.exp(-0.02*(age[i]-start_age))*(ct5r1[i])**(1-9))/(1-9)
    e13=(math.exp(-0.02*(age[i]-start_age))*(ct5r1[i])**(1-10))/(1-10)
    
    a5_t5r1.append(e1)
    a5_t5r2.append(e2)
    a5_t5r3.append(e3)
    a5_t5r4.append(e4)
    a5_t5r5.append(e5)

    a5_t2r2.append(e6)
    a5_t3r2.append(e7)
    a5_t4r2.append(e8)
    a5_t6r2.append(e9)
    a5_t7r2.append(e10)
    a5_t8r2.append(e11)
    a5_t9r2.append(e12)
    a5_t10r2.append(e13)

#Creating expected Utility, as a sum of the utility gained in each period
Eutilt5r1=sum(a5_t5r1)
Eutilt5r2=sum(a5_t5r2)
Eutilt5r3=sum(a5_t5r3)
Eutilt5r4=sum(a5_t5r4)
Eutilt5r5=sum(a5_t5r5)

Eutilt2r2=sum(a5_t2r2)
Eutilt3r2=sum(a5_t3r2)
Eutilt4r2=sum(a5_t4r2)
Eutilt6r2=sum(a5_t6r2)
Eutilt7r2=sum(a5_t7r2)
Eutilt8r2=sum(a5_t8r2)
Eutilt9r2=sum(a5_t9r2)
Eutilt10r2=sum(a5_t10r2)
```

```python
#Calculating Certainty Equivalent for different values of theta and rho
ce_t5r1=(Eutilt5r1*(1-5)/Funcion_g_ce(5, 0.01))**(1/(1-5))
ce_t5r2=(Eutilt5r2*(1-5)/Funcion_g_ce(5, 0.02))**(1/(1-5))
ce_t5r3=(Eutilt5r3*(1-5)/Funcion_g_ce(5, 0.03))**(1/(1-5))
ce_t5r4=(Eutilt5r4*(1-5)/Funcion_g_ce(5, 0.04))**(1/(1-5))
ce_t5r5=(Eutilt5r5*(1-5)/Funcion_g_ce(5, 0.05))**(1/(1-5))

ce_t2r2=(Eutilt2r2*(1-2)/Funcion_g_ce(2, 0.02))**(1/(1-2))
ce_t3r2=(Eutilt3r2*(1-3)/Funcion_g_ce(3, 0.02))**(1/(1-3))
ce_t4r2=(Eutilt4r2*(1-4)/Funcion_g_ce(4, 0.02))**(1/(1-4))
ce_t6r2=(Eutilt6r2*(1-6)/Funcion_g_ce(6, 0.02))**(1/(1-6))
ce_t7r2=(Eutilt7r2*(1-7)/Funcion_g_ce(7, 0.02))**(1/(1-7))
ce_t8r2=(Eutilt8r2*(1-8)/Funcion_g_ce(8, 0.02))**(1/(1-8))
ce_t9r2=(Eutilt9r2*(1-9)/Funcion_g_ce(9, 0.02))**(1/(1-9))
ce_t10r2=(Eutilt10r2*(1-10)/Funcion_g_ce(10, 0.02))**(1/(1-10))

#Calculating Welfare Loss
WL_t5r1=np.array((ce_t5r2-ce_t5r1)/ce_t5r1)

WL_t5r2=(ce_t5r2-ce_t5r2)/ce_t5r2
WL_t5r3=(ce_t5r2-ce_t5r3)/ce_t5r3
WL_t5r4=(ce_t5r2-ce_t5r4)/ce_t5r4
WL_t5r5=(ce_t5r2-ce_t5r5)/ce_t5r5

WL_t2r2=(ce_t5r2-ce_t2r2)/ce_t2r2
WL_t3r2=(ce_t5r2-ce_t3r2)/ce_t3r2
WL_t4r2=(ce_t5r2-ce_t4r2)/ce_t4r2
WL_t6r2=(ce_t5r2-ce_t6r2)/ce_t6r2
WL_t7r2=(ce_t5r2-ce_t7r2)/ce_t7r2
WL_t8r2=(ce_t5r2-ce_t8r2)/ce_t8r2
WL_t9r2=(ce_t5r2-ce_t9r2)/ce_t9r2
WL_t10r2=(ce_t5r2-ce_t10r2)/ce_t10r2
```
